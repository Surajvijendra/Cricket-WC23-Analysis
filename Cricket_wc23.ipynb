```python
# importing necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
team_data = pd.read_csv(r"E:\Data scince\Machine learning\Mentorness\CWC23_all_innings.csv")
team_data
```

# EDA

```python
team_data.isnull().sum()
```

```python
# Replacing the null values with 0

team_data.fillna(0,inplace=True)
pd.set_option('display.max.rows',1408)
team_data
```

```python
# overall bowling performence team wise
bowlers_data = team_data[team_data['bat_or_bowl'] == 'bowl'].groupby('team').agg({
'bb_bf':'sum',
'runs':'sum',
'wkts':'sum',
'wicketball_prob':'mean',
'runs_per_ball':'mean',
'overs':'sum',
'mdns':'sum',
'econ':'mean',
'4s':'sum',
'6s':'sum','sr':'mean','not_out':'sum','mins':'sum'})
bowlers_data.reset_index(inplace=True)
bowlers_data
```

```python
wkts_sorted_values = bowlers_data.sort_values(['wkts'], ascending=False)

overs_sorted_values = bowlers_data.sort_values(['overs'], ascending=False)


econ_sorted_values = bowlers_data.sort_values(['econ'], ascending=True)


mdns_sorted_values = bowlers_data.sort_values(['mdns'], ascending=False)

print(wkts_sorted_values[['team','wkts','overs','econ','mdns']])
```

```python
# plot for bowlers

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

sns.barplot(x='team', y='wkts', data=wkts_sorted_values, ax=axes[0, 0])
axes[0, 0].set_title('Wickets taken by teams')

sns.barplot(x='team', y='overs', data=overs_sorted_values, ax=axes[0, 1])
axes[0, 1].set_title('Overs delivered by teams')

sns.barplot(x='team', y='econ', data=econ_sorted_values, ax=axes[1, 0])
axes[1, 0].set_title('Team with best Economy')

sns.barplot(x='team', y='mdns', data=mdns_sorted_values, ax=axes[1, 1])
axes[1, 1].set_title('Maiden Overs bowled by teams')

plt.show()
```

```python
# over all batting performence team wise
batters_data = team_data[team_data['bat_or_bowl'] == 'bat'].groupby('team').agg({
'bb_bf':'sum',
'runs':'sum',
'wkts':'sum',
'wicketball_prob':'mean',
'runs_per_ball':'mean',
'overs':'sum',
'mdns':'sum',
'econ':'mean',
'4s':'sum',
'6s':'sum','sr':'mean','not_out':'sum','mins':'sum'})
batters_data.reset_index(inplace=True)
batters_data
```

```python
# plot for batters


fig,axes=plt.subplots(nrows=2, ncols=2, figsize=(12,10))

sns.barplot(x='team',y='runs',data=runs_sorted_values, ax=axes[0,0])
axes[0,0].set_title('Top Runs scored by Teams')

sns.barplot(x='team',y='sr',data=sr_sorted_values,ax=axes[0,1])
axes[0,1].set_title('Team with best strike rate')

sns.barplot(x='team',y='4s',data=max_4s,ax=axes[1,0])
axes[1,0].set_title('Team with most 4s')

sns.barplot(x='team',y='6s',data=max_6s, ax=axes[1,1])
axes[1,1].set_title('Team with most 6s')

plt.tight_layout()
plt.show()

sns.barplot(x='team',y='mins',data=mins)
plt.title('Team wise Players who Spent more time while batting')
plt.show()
```
